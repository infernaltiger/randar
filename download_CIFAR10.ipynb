```python
import torch
from torchvision import datasets, transforms
```

```python
transform = transforms.ToTensor()

# Загрузка обучающего и тестового наборов
train_dataset = datasets.CIFAR10(
    root='RandAR\\data\\cifar10',          # куда сохранить
    train=True,
    download=True,          # автоматически скачает, если нет
    transform=transform
)

test_dataset = datasets.CIFAR10(
    root='RandAR\\data\\cifar10',
    train=False,
    download=True,
    transform=transform
)

# Пример: посмотреть первое изображение 
image, label = train_dataset[0]
print("Image shape:", image.shape)  # torch.Size([3, 32, 32])
print("Label:", label, "Class:", train_dataset.classes[label])
```
